```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Generate synthetic data
np.random.seed(42)

# Generate synthetic dataset
x = np.random.uniform(0, 5, 200)
noise = np.random.normal(0, 1, 200)
y = 3 + 4*x + noise

# Add bias column
X = np.c_[np.ones(len(x)), x]

print("Dataset generated successfully.")

```

```python
# Closed-form solution (Normal Equation)
plt.figure()
plt.scatter(x, y)
plt.title("Raw Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

```python
# Gradient Descent
theta_closed = np.linalg.inv(X.T @ X) @ X.T @ y

print("Closed-Form Solution")
print("Intercept:", theta_closed[0])
print("Slope:", theta_closed[1])

```

```python
# Plot data and fitted lines
plt.figure()
plt.scatter(x, y)
plt.plot(x, X @ theta_closed)
plt.title("Closed-Form Fitted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

```python
# Plot loss vs iterations
theta = np.zeros(2)
learning_rate = 0.05
iterations = 1000
m = len(y)

loss_history = []

for i in range(iterations):
    predictions = X @ theta
    errors = predictions - y
    mse = np.mean(errors**2)
    loss_history.append(mse)
    gradient = (2/m) * X.T @ errors
    theta -= learning_rate * gradient

print("Gradient Descent Solution")
print("Intercept:", theta[0])
print("Slope:", theta[1])

```

```python
plt.figure()
plt.scatter(x, y)
plt.plot(x, X @ theta)
plt.title("Gradient Descent Fitted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.figure()
plt.scatter(x, y)
plt.plot(x, X @ theta_closed)
plt.plot(x, X @ theta)
plt.title("Closed-Form vs Gradient Descent")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.figure()
plt.plot(loss_history)
plt.title("Loss vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.show()
```
